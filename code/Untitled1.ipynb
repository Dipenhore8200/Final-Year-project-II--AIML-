```python
st = set()
def extract_uniques(row):
#     print(row['Entry'], end=' ')

#     st.add(row['Entry'])
    print(row.shape[0])
    
    for i in range(row.shape[0]):
        st.add(row[i])
```

```python
new_df = df[~df['Interacts with'].isnull()]
```

```python
new_df['Interacts with'].iloc[5]
```

```python
st2 = set()
def extract_uniques2(row):
    print(row.shape[0])
    sum=0
    for i in range(row.shape[0]):
        l=[]
        proteins = row.iloc[i]
        if ';' in proteins:
            l = proteins.split("; ")
        else:
            l.append(proteins);
            
        sum+=len(l)
        
        for i in l:
            st2.add(i)
            
    print(sum)
        
```

```python
extract_uniques2(new_df['Interacts with'])
```

```python
len(st2)
```
